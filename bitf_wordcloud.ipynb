```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
import matplotlib as mpl
```

```python
# Read CSV File
import os
# df = pd.read_csv("desktop\Word_Cloud_NVDA.csv")
# read the whole text
text = open('bitf_text.rtf', encoding="utf8").read()

comment_words = ""
stopwords = set(STOPWORDS)
```

```python
# read the mask/color image taken from
#coloring = np.array(Image.open('.jpg'))
stopwords = set(STOPWORDS)
custom_stop_words = [""]
[stopwords.add(n) for n in custom_stop_words]
```

```python
wc = WordCloud(background_color="white" , width=500, max_words=3500, 
              stopwords=stopwords, max_font_size=100, random_state=42)

#generate word cloud
wc_image= wc.generate(text)
```

```python
mpl.rcParams["figure.figsize"]=[20.0,10.0]



plt.imshow(wc_image)
plt.axis("off")
fontdict={"fontsize": 48, "fontweight":"bold"}
plt.title("BITF Word Cloud", fontdict=fontdict)
plt.show()
```
